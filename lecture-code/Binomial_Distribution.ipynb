```python
import matplotlib.pyplot as plt
from scipy.stats import binom
```

```python
# Parameters for the binomial distribution (number of coin flips and probability of success)
n_values = [10, 100, 200]  # Different numbers of coin flips
p = 0.5  # Probability of getting heads (success)
```

```python
plt.figure(figsize=(12, 6))  # Create a larger figure

# Plotting the binomial distributions for different n values
for n in n_values:
    x = range(n + 1)  # Possible number of successes (0 to n)
    binomial_dist = binom.pmf(x, n, p)  # Probability Mass Function
    plt.plot(x, binomial_dist, label=f'n = {n}')

plt.title('Binomial Distributions of Coin Flips')
plt.xlabel('Number of Successes (k)')
plt.ylabel('Probability')
plt.xticks(range(0, max(n_values) + 1, max(n_values) // 10))
plt.legend()
plt.grid(axis='y')
plt.show()
```

```python
plt.figure(figsize=(12, 6))  # Create a larger figure

# Plotting the cumulative distribution functions for different n values
for n in n_values:
    x = range(n + 1)  # Possible number of successes (0 to n)
    binomial_dist = binom.pmf(x, n, p)
    k_values = [k - n/2 for k in x]  # k - n/2
    plt.plot(k_values, binomial_dist, label=f'n = {n}')

plt.title('Binomial Distributions of Coin Flips')
plt.xlabel('k - n/2')
plt.ylabel('Probability')
plt.legend()
plt.grid()
plt.legend()
plt.show()
```
