# Zero Correlation

How does correlation look when it's symmetric along an axis?

<img src="imgs/zero-correlation.png" alt="Zero Correlation" width="500" />

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Symmetric along the x-axis

```python
# Create an empty DataFrame
df = pd.DataFrame()

# Create values for feature X ranging from -2 to +2 with 100 samples
df["X"] = np.random.uniform(-2, 2, 100)

# Create Y values based on a quadratic relationship Y = X^2
df["Y"] = df["X"] ** 2

# Create error values ranging from -0.5 to 0.5 with the same number of samples
df["error"] = np.random.uniform(-0.5, 0.5, 100)
```

```python
# Plot X and Y on the same graph
plt.figure(figsize=(8, 6))

plt.scatter(df["X"], df["Y"], color="blue", alpha=0.5)
plt.title("X vs Y (Quadratic Relationship)")
plt.xlabel("X")
plt.ylabel("Y = X^2")
plt.grid(True)

plt.show()
```

```python
# Calculate covariance between X and Y
covariance_xy = np.cov(df["X"], df["Y"], ddof=0)[0, 1]

# Calculate standard deviations of X and Y
std_dev_x = np.std(df["X"], ddof=0)
std_dev_y = np.std(df["Y"], ddof=0)

print("Covariance between X and Y:", covariance_xy)
print("Standard deviation of X:", std_dev_x)
print("Standard deviation of Y:", std_dev_y)

# Calculate correlation coefficient
correlation_coefficient = covariance_xy / (std_dev_x * std_dev_y)

print("Correlation coefficient between X and Y:", correlation_coefficient)
```

* Correlation is almost zero.
* Now let's try the same thing with more data points

```python
# Create an empty DataFrame
df = pd.DataFrame()

# Create values for feature X ranging from -2 to +2 with 10000 samples
df["X"] = np.random.uniform(-2, 2, 10000)

# Create Y values based on a quadratic relationship Y = X^2
df["Y"] = df["X"] ** 2

# Create error values ranging from -0.5 to 0.5 with the same number of samples
df["error"] = np.random.uniform(-0.5, 0.5, 10000)
```

```python
# Calculate covariance between X and Y
covariance_xy = np.cov(df["X"], df["Y"], ddof=0)[0, 1]

# Calculate standard deviations of X and Y
std_dev_x = np.std(df["X"], ddof=0)
std_dev_y = np.std(df["Y"], ddof=0)

print("Covariance between X and Y:", covariance_xy)
print("Standard deviation of X:", std_dev_x)
print("Standard deviation of Y:", std_dev_y)
```

```python
# Calculate correlation between X and Y
correlation_xy = covariance_xy / (std_dev_x * std_dev_y)

print("Correlation between X and Y:", correlation_xy)
```

Correlation is closer to zero.
